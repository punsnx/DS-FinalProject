## Data Cleaning

### เลือกคอลัมน์ที่ต้องการที่จะนำมาวิเคราะห์

```python
df = df[['id','release_date','revenue','genres','vote_count','runtime','budget','vote_average']]
```

### ลบค่าที่เป็นค่าว่างออก และค่าซ้ำออก

```python
df = df.dropna(subset=['release_date','revenue','genres'])
df = df.drop_duplicates(subset=['id'])
```

### แปลง budget ให้เป็นค่าตัวเลข

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
```

### เลือกเฉพาะภาพยนต์ที่มี revenue > 0

```python
df = df[df['revenue'] > 0]
df['runtime'].fillna(df['runtime'].median(), inplace=True)
```

```python
df.info()
```

```python
df.head()
```

### แปลงค่า release_date ให้เป็น datetime

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_month'] = df['release_date'].dt.month
df['release_year'] = df['release_date'].dt.year
df.head()
```

### แปลง Genres ให้อยู่ในรูปที่เข้าถึงได้ง่าย

```python

df.genres = df.genres.apply(eval).apply(lambda x: [genre['name'] for genre in x] if x else pd.NA)
df=df.dropna(subset=['genres'])
df['genres'] = df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))
df['genres']
```
