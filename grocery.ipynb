```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sale_df = pd.read_csv("datasets/DS-GroceryDataset/sales.csv")
product_df = pd.read_csv("datasets/DS-GroceryDataset/products.csv")
```

```python
sale_df
```

```python
product_df
```

```python
sale_df = sale_df.join(product_df.set_index('ProductID'), on='ProductID', how='left', lsuffix='_sale', rsuffix='_product')
sale_df
```

```python
category_df = pd.read_csv("datasets/DS-GroceryDataset/categories.csv")
category_df
```

```python
sale_df = sale_df.join(category_df.set_index('CategoryID'), on='CategoryID', how='left', lsuffix='_sale', rsuffix='_category')
sale_df
```

```python
sale_df.TotalPrice = sale_df.Price *sale_df.Quantity
sale_df
```

```python
customer_df = pd.read_csv("datasets/DS-GroceryDataset/customers.csv")
customer_df
```

```python
city_df = pd.read_csv("datasets/DS-GroceryDataset/cities.csv")
city_df
```

```python
country_df = pd.read_csv("datasets/DS-GroceryDataset/countries.csv")
country_df
```

```python
city_df = city_df.join(country_df.set_index('CountryID'), on='CountryID', how='left', lsuffix='_city', rsuffix='_country')
city_df
```

```python
customer_df = customer_df.join(city_df.set_index('CityID'), on='CityID', how='left', lsuffix='_customer', rsuffix='_city')
customer_df
```

```python
employee_df = pd.read_csv("datasets/DS-GroceryDataset/employees.csv")
employee_df
```

```python
employee_df = employee_df.join(city_df.set_index("CityID"),on="CityID",how="left",lsuffix="_employee",rsuffix="_city")
employee_df
```

```python
sale_df = sale_df.join(customer_df.set_index("CustomerID"),on="CustomerID",how="left",lsuffix="_sale",rsuffix="_customer")
sale_df
```

```python
sale_df = sale_df.join(employee_df.set_index("EmployeeID"),on="SalesPersonID",how="left",lsuffix="_customer",rsuffix="_employee")
sale_df
```

```python
sale_df = sale_df[['SalesID','SalesDate','TransactionNumber', 'CustomerID','FirstName_customer','MiddleInitial_customer','LastName_customer',
                   'CountryID_customer','CountryCode_customer', 'CountryName_customer', 'CityID_customer',
                   'CityName_customer','Address','Zipcode_customer','ProductID','ProductName','Quantity', 
                   'Price','TotalPrice','Discount','CategoryID','CategoryName','Class','Resistant', 'IsAllergic', 'VitalityDays',
                   'SalesPersonID','FirstName_employee', 'MiddleInitial_employee', 'LastName_employee','BirthDate', 'Gender','HireDate',
                   'CountryID_employee','CountryCode_employee','CountryName_employee','CityID_employee','CityName_employee','Zipcode_employee',
                   'ModifyDate'
                   ]]
```
