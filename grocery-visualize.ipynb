```python
import pandas as pd 
import numpy as numpy
import seaborn as sns
```

```python
sale_df = pd.read_csv("datasets/Grocery-sales_joined.csv")
sale_df
```

```python
sale_df.SalesDate = pd.to_datetime(sale_df['SalesDate'])
sale_df
```

```python
sale_df.reset_index(inplace=True)
sale_df.set_index("SalesDate",inplace=True)
sale_df.sort_index(ascending=True,inplace=True)
sale_df
```

```python
sale_date_item = sale_df.resample("D").Quantity.sum()
```

```python
sale_date_item_df = sale_date_item.to_frame()

# Plotting the sales quantity over time
sale_date_item_df.plot(y='Quantity', kind='line', title='Daily Sales Quantity', xlabel='Date', ylabel='Quantity')
sale_date_item_df
```
